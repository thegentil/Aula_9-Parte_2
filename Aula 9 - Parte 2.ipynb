**2.2) Forma analítica (expressão) da distribuição:**

> Função densidade de probabilidade (para $x>0$): $$f(x) = \frac{\beta^{-\alpha} x^{\alpha - 1} e^{- \frac{x}{\beta}}}{\Gamma(\alpha)}$$ 

> Onde, a função gama completa é: $$ \Gamma(\alpha) = \int^{1}_{0} t^{\alpha - 1} e^{-t} dt$$

**2.3) Parâmetros**

> Os parâmetros são:

> $\alpha$: Parâmetro de forma

> $\beta$: Parâmetro de escala

**2.4) Fórmula específica para calcular esperança (média) e variáncia em função dos parâmetros:**

> Média:
> $\alpha \beta$

> Variância:
> $\alpha \beta^2$
